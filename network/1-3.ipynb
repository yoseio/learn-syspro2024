```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

udp = pd.read_csv("1-3-udp.csv")
tcp = pd.read_csv("1-3-tcp.csv")
```

```python
plt.figure(figsize=(10, 5))

plt.scatter("i", "rtt_us", data=udp, label="UDP")

plt.grid(True)
plt.xlabel("Data (2 ^ x bytes)")
plt.ylabel("RTT (us)")
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.scatter("i", "rtt_us", data=tcp, label="TCP")

plt.grid(True)
plt.xlabel("Data (2 ^ x bytes)")
plt.ylabel("RTT (us)")
plt.legend()

plt.show()
```

```python
udp_mean = udp.groupby("i", as_index=False)["rtt_us"].mean()
tcp_mean = tcp.groupby("i", as_index=False)["rtt_us"].mean()
```

```python
plt.figure(figsize=(10, 5))

plt.plot("i", "rtt_us", data=udp_mean, label="UDP (mean)", marker='.')
plt.plot("i", "rtt_us", data=tcp_mean, label="TCP (mean)", marker='.')

plt.grid(True)
plt.xlabel("Data (2 ^ x bytes)")
plt.ylabel("RTT (us)")
plt.xticks(np.arange(0, 21, 1))
plt.legend()

plt.show()
```
