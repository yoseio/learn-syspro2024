```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def mm1(λ, μ):
    assert (λ < μ).all()
    rho = λ / μ
    return rho / (1 - rho) * (1 / μ)

plt.figure(figsize=(10, 5))
μ = np.arange(2.0, 3.0, 0.1)

data = pd.read_csv("4-4-1.csv")
plt.plot(μ, mm1(0.5, μ), label="Theoretical, λ = 0.5", marker="o", markersize=10, lw=5)
plt.plot("μ", "loss", data=data, label="Simulation, λ = 0.5", marker="o")

data = pd.read_csv("4-4-2.csv")
plt.plot(μ, mm1(1.0, μ), label="Theoretical, λ = 1.0", marker="x", markersize=10, lw=5)
plt.plot("μ", "loss", data=data, label="Simulation, λ = 1.0", marker="x")

data = pd.read_csv("4-4-3.csv")
plt.plot(μ, mm1(1.5, μ), label="Theoretical, λ = 1.5", marker="^", markersize=10, lw=5)
plt.plot("μ", "loss", data=data, label="Simulation, λ = 1.5", marker="^")

plt.grid(True)
plt.xlabel(r"$μ$")
plt.ylabel(r"$Loss rate$")
plt.legend()

plt.show()
```
