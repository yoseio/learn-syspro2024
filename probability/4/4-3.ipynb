```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(10, 5))

sym = ["o", "s", "D", "x", "^"]

for s in range(5):
  data = pd.read_csv(f"4-3-{s+1}.csv")
  plt.plot("", "loss", data=data, label=f"S = {s+1}", marker=sym[s])

plt.grid(True)
plt.xlabel(r"$$")
plt.ylabel(r"$Loss rate$")
plt.ylim(0, 1)
plt.legend()

plt.show()
```
