```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("4-2-1.csv")

λ = data["λ"].to_numpy()
μ = data["μ"].to_numpy()
theoretical = λ / (λ + μ)

plt.figure(figsize=(10, 5))

plt.plot(μ, theoretical, label="Theoretical", marker="x", markersize=10, lw=5)
plt.plot("μ", "loss", data=data, label="Simulation", marker="o")

plt.grid(True)
plt.xlabel(r"$μ$")
plt.ylabel(r"$Loss rate$")
plt.ylim(0, 1)
plt.legend()

plt.show()
```
