```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


data = pd.read_csv("3-3.csv")
k = 7


def draw(λ):
  plt.figure(figsize=(5, 5))

  poisson = np.random.poisson(λ, 10000)
  x = np.arange(0, k)
  y = [np.sum(poisson == i) / len(poisson) for i in range(k)]
  plt.plot(x, y, label=f"True, λ = {λ}", marker="o", markersize=10, lw=5)

  plt.plot("i", str(λ), data=data, label="Simulation", marker="o")

  plt.grid(True)
  plt.xlabel(r"$k$")
  plt.ylabel(r"$P(k)$")
  plt.xlim(0, 6)
  plt.ylim(0, 1)
  plt.legend()

  plt.show()
```

```python
draw(1.0)
```

```python
draw(1.5)
```

```python
draw(2.0)
```

```python
draw(2.5)
```
