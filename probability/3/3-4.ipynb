```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("3-4.csv")

def plot(λ):
  plt.figure(figsize=(15, 5))

  plt.plot("m", str(λ), data=data, label=f"λ = {λ}")

  plt.grid(True)
  plt.xlabel(r"$m$")
  plt.ylabel(r"$w_m$")
  plt.legend()

  plt.show()
```

```python
plot(1.0)
```

```python
plot(1.5)
```

```python
plot(2.0)
```

```python
plot(2.5)
```
