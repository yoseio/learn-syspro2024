```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv('1-4.csv')

theoretical = sum([i * p for i, p in zip([1, 2, 3, 4, 5, 6], [1/9, 2/9, 1/9, 2/9, 1/9, 2/9])])

plt.figure()

plt.plot("n", "mean", data=data, label='Actual')
plt.axhline(y=theoretical, color='r', linestyle='--', label='Theoretical')

plt.grid(True)
plt.legend()

plt.show()
```

```python
sigma = np.sqrt(sum([i ** 2 * p for i, p in zip([1, 2, 3, 4, 5, 6], [1/9, 2/9, 1/9, 2/9, 1/9, 2/9])]) - theoretical**2)
vals = np.arange(1, 1001)
errors = 1.96 * sigma / np.sqrt(vals)
upper = theoretical + errors
lower = theoretical - errors

plt.figure()

plt.plot("n", "mean", data=data, label='Actual')
plt.axhline(y=theoretical, color='r', linestyle='--', label='Theoretical')
plt.fill_between(vals, lower, upper, color='gray', alpha=0.3, label='±2 SE')

plt.grid(True)
plt.legend()

plt.show()
```
