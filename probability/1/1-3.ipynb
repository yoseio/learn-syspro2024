```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

theoretical = 3.5
data = pd.read_csv('1-3.csv')

plt.figure()

plt.plot("n", "mean", data=data, label='Actual')
plt.axhline(y=theoretical, color='r', linestyle='--', label='Theoretical')

plt.grid(True)
plt.legend()

plt.show()
```

```python
sigma = np.sqrt((6**2-1)/12)
vals = np.arange(1, 1001)
errors = 1.96 * sigma / np.sqrt(vals)
upper = theoretical + errors
lower = theoretical - errors

plt.figure()

plt.plot("n", "mean", data=data, label='Actual')
plt.axhline(y=theoretical, color='r', linestyle='--', label='Theoretical')
plt.fill_between(vals, lower, upper, color='gray', alpha=0.3, label='2 SE')

plt.grid(True)
plt.legend()

plt.show()
```
