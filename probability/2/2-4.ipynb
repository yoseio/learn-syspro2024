```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("2-4.csv")
```

```python
plt.figure(figsize=(5, 5))

plt.scatter("mt_x", "mt_y", data=data, c="red")

plt.grid(True)
plt.xlabel(r"$X_i$")
plt.ylabel(r"$X_{i+1}$")
plt.legend()

plt.show()
```
