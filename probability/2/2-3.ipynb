```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("2-3.csv")

plt.figure(figsize=(10, 5))

plt.plot("t", "st", data=data, c="red", linewidth=0.7)

plt.grid(True)
plt.xlabel(r"$t$")
plt.ylabel(r"$S_t$")

plt.show()
```
