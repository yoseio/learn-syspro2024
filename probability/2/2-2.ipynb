```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv("2-2.csv")
n = 10000

def draw(λ):
    plt.figure(figsize=(5, 5))

    x = np.linspace(0, 5, n)
    y = λ * np.exp(-λ * x)
    plt.plot(x, y, label=f"True, λ = {λ}", c="red")

    bins = np.arange(0, 5, 0.2)
    plt.hist(str(λ), data=data, bins=bins, density=True, label="Random values", alpha=0.5, edgecolor='black')

    plt.grid(True)
    plt.xlabel("x")
    plt.ylabel("Probability density")
    plt.xlim(0, 5)
    plt.ylim(0, 2)
    plt.yticks(np.arange(0.0, 2.1, 0.5))
    plt.legend()

    plt.show()
```

```python
draw(1.0)
```

```python
draw(1.5)
```

```python
draw(2.0)
```

```python
import scipy.stats as stats


def v_se2(λ):
  n = 10000
  v = 1 / (λ ** 2)
  lower = (n - 1) * v / stats.chi2.ppf(0.975, n - 1)
  upper = (n - 1) * v / stats.chi2.ppf(0.025, n - 1)
  return (lower, upper)


print(v_se2(1.0))
print(v_se2(1.5))
print(v_se2(2.0))
```
